## Q1) Write a python program to create a doubly linked list of 10 elements and Delete the alternate nodes in it.

## Q2) Check if a given DLL is palindrome or not.

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.prev = None
        self.next = None

class Doubly_Linked_List:
    def __init__(self):
        self.head = None

    def Insert(self, data):
        newNode = Node(data)
        if self.head:
            current = self.head
            while current.next != None:
                current = current.next
            current.next = newNode
            newNode.prev = current
        else:
            self.head = newNode

    def Print_DLL(self):
        current = self.head
        while current:
            print(current.data, end=" <=> ")
            current = current.next
        print("Null")

    # here, we have make a function for deleting alternate nodes
    def deleteOddNodes(self):
        # now to delete a node from DLL,
        # it should contain atleast one node
        if (self.head != None):
            """
            Here,we are maintaining two pointers (i.e. oddNode and evenNode)
            So here, oddNode is pointing to head node and evenNode is pointing to next node after head node
            """
            oddNode = self.head
            evenNode = self.head.next 

            """Now we are running a while loop until anyone one of these nodes is none"""
            while(evenNode != None and oddNode != None):
                """
                Now here is little bit tricky part, 
                we are giving address of next of evenNode to next of oddNode,
                
                (Now situation is like this, suppose we have total of 3 nodes in our DLL,
                and is it marked as 1, 2 and 3, so what this step is doing, 
                it is forming a link between 1st and 3rd node, why?
                because 1st node is oddNode and 2nd node is evenNode 
                and we are giving address of next of evenNode(i.e. 3rd node) to next of oddNode,
                so now next of oddNode will be 3rd node not 2nd node)
                
                So by this we have simply connected 1st node to 3rd node 
                or we can say that we have deleted the 2nd node existance
                """
                oddNode.next = evenNode.next
                evenNode = None
                oddNode = oddNode.next
                if(oddNode != None):
                    evenNode = oddNode.next
        else:
            print("Given DLL is Empty.")

    # here, we have made a function, for checking palindrome in DLL
    def Palindrome(self):
        """
        here, we are simply maintaining a list, in which we are storing ever value of DLL
        and then checking if this list is equal to reverse of this list,
        if yes, then is it will be palindrome otherwise not
        """
        current = self.head
        l = []
        while current:
            l.append(current.data)
            current = current.next
        l1 = l[::-1]
        if l==l1:
            print("Given DLL is Palindrome")
        else:
            print("Given DLL is not Palindrome")

    def Length(self):
        current = self.head
        a = 1
        while current.next != None:
            current = current.next
            a = a + 1
        print(f"Length of our Doubly Linked list is {a}")

dll = Doubly_Linked_List()
dll.Insert(1)
dll.Insert(2)
dll.Insert(3)
dll.Insert(4)
dll.Insert(5)
dll.Insert(4)
dll.Insert(3)
dll.Insert(2)
dll.Insert(1)
print("Given Doubly linked list is :")
dll.Print_DLL()
print(" ")
print(" ")
dll.Length()
dll.deleteOddNodes()
print(" ")
print(" ")
print("After Deleting Alternate Nodes ")
dll.Print_DLL()
print(" ")
print(" ")
dll.Length()
print(" ")
print(" ")
dll.Palindrome()
```
