## Q2) Write a python program to reverse the nodes of existing linked list.

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def Insert(self, data):
        newNode = Node(data)
        if self.head:
            current = self.head
            while current.next:
                current = current.next
            current.next = newNode
        else:
            self.head = newNode

    def PrintLL(self):
        current = self.head
        while current:
            print(current.data, end=" => ")
            current = current.next
        print("Null")

    # Now we have made a function which will print
    # the values in odd nodes of our LL
    def Print_OddNodes(self):
        """
        Now to print the value in odd nodes,
        We have to maintain one pointer 
        (i.e. here we are using 'current' which is pointing towards head initially), 
        and a counter (i.e. here we have 'i' for counting the nodes)
        """
        current = self.head
        i = 0
        while i>=0 and current:
            """
            Now if my counter (i.e. 'i') doesn't gives zero as remainder when we take its modulo with 2, 
            that means, this current node is odd node and we will print its value (i.e. using current.data)
            """
            if (i%2 != 0):
                print(current.data, end=" => ")
            current = current.next
            i = i + 1
        print("Null")

    
    # Now using same concept discussed in previous function,
    # we have made another function which will print the values in even nodes
    def Print_EvenNodes(self):
        current = self.head
        i = 0
        while i>=0 and current:
            if (i%2 == 0):
                print(current.data, end=" => ")
            current = current.next
            i = i + 1
        print("Null")

    
    # here, we have made function which will reverse the order of our current LL
    def Reverse_order(self):
        """
        Now here we have to maintain two pointers,
        (i.e. one for storing value of current node and other for storing value of previous node)
        """
        prev = None
        current = self.head
        while current is not None:
            """
            here is the complex part comes, when we have to reverse the order, 
            we will reverse the order, one node at a time
            So here we are giving current.next pointer address to 'next' pointer
            and at last two steps (i.e. in prev = current and current = next) 
            we are just reversing the role of both the nodes
            (i.e. current node and next of current node)
            """
            next = current.next
            current.next = prev
            prev = current
            current = next
        """
        So after completing while loop the last node will we stored in 'prev' pointer
        So we are just making the last node as head of our current LL
        """
        self.head = prev  


    """
    At last we are making Length function, just to make sure,
    while printing the values of odd or even nodes, 
    we not deleting anything from LL 
    (i.e. we want to make sure we are just printing values,
    not deleting any value while printing)
    """
    def Length(self):
        current = self.head
        a = 1
        while current.next != None:
            current = current.next
            a = a + 1
        print(f"Length of our Singly Linked list is {a}")

ll = LinkedList()
ll.Insert(1)
ll.Insert(2)
ll.Insert(3)
ll.Insert(4)
ll.Insert(5)
ll.Insert(6)
ll.Insert(7)
ll.Insert(8)
ll.Insert(9)
ll.Insert(10)
print("Given Singly linked list is : ")
ll.PrintLL()
print(" ")
print(" ")
ll.Length()
print(" ")
print(" ")
print("Elements in Odd Nodes are :")
ll.Print_OddNodes()
print(" ")
print(" ")
ll.Length()
print(" ")
print(" ")
print("Elements in Even Nodes are :")
ll.Print_EvenNodes()
print(" ")
print(" ")
ll.Length()
print(" ")
ll.Reverse_order()
print(" ")
print("Reverse order of our linked list is : ")
ll.PrintLL()
print(" ")
print(" ")
ll.Length()
```
