# Q1) Create a BST having following features:
## i) insert ii) Delete iii) find particular value 
## iv) find max and min value v) find height

```python
#                50                                         50                                      50
#             /     \        After deleting 40 & 55      /      \        After deleting 45         /  \
#           40        60    <----------------------->   45        60      <------------->        30    60
#          /  \      /  \                              /           \                                    
#        30    45   55   70                           30            70                                  
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


def Insert_value(root, data):
    if root.data: # cheking for root node 
        if data < root.data: # if data is less then root node then it will be added in left side of tree
            if root.left is None:
                root.left = Node(data)
            else:
                Insert_value(root.left, data) # recursively calling the function in left side

        elif data > root.data: # if data is greater then root node then it will be add in rightside tree
            if root.right is None:
                root.right = Node(data)
            else:
                Insert_value(root.right, data) # recursively calling the function in right side
    else:
        root.data = data


def find_value(root, value):
    if value < root.data: # if value is less then root value then it we will search in left side of tree
        if root.left is None:
            return (f"{value} Not found")
        return find_value(root.left, value) # recursively calling the function in left side tree

    elif value > root.data:
        if root.right is None:
            return (f"{value} Not found")
        return find_value(root.right, value)
    
    else: # if value is equal to root value then return root value found
        return (f"{value} is found")


def Max_value(root):
    current = root
    while current.right is not None: # searching in right side tree for getting max value in tree
        current = current.right

    return current.data

def Min_value(root):
    current = root
    while current.left is not None: # searching in right side tree for getting max value in tree
        current = current.left

    return current.data

def Delete_Value(root, value):
    if root.data is None: # if empty tree is there
        return root

    elif  value < root.data: # if value is less then root value then it we will go in left side of tree
        root.left = Delete_Value(root.left, value) # recursively calling in left tree 

    elif value > root.data:
        root.right = Delete_Value(root.right, value)

    else : # means we found the value to be deleted i.e. value = root.data

        # case 1 : when root to be deleted has no children 
        if (root.left is None) and (root.right is None): 
            del root
            root = None
            return root

        # case 2 : when root to be deleted has 1 child 
        if root.left is None: # if there is no left children
            temp = root.right
            root = None
            return temp

        elif root.right is None: # if there is not right children
            temp = root.left
            root = None
            return temp

        # case 3 : when root to be deleted has two children
        temp = Min_value(root.right) 

        root.data = temp  # Copy the inorder successor's content to this node
        
        root.right = Delete_Value(root.right, temp) # Delete the inorder successor

    return root


def Height(root):
    if root is None: # if empty tree 
        return -1
    else: # calling this function in left and right tree and return that value which has max value
        return (1 + max(Height(root.left), Height(root.right)))


def Pre_Ordered(cur): # Root ---> Left ---> Right
    if cur != None:
        print(cur.data, end=" ")
        Pre_Ordered(cur.left)
        Pre_Ordered(cur.right)
       
def Post_Ordered(cur): # Left ---> Right ---> Root
    if cur != None:
        Post_Ordered(cur.left)
        Post_Ordered(cur.right)
        print(cur.data, end=" ")

def In_Ordered(cur): # Left ---> Root ---> Right
    if cur != None:
        In_Ordered(cur.left)
        print(cur.data, end=" ")
        In_Ordered(cur.right)

if __name__=="__main__":
    root = Node(50)
    Insert_value(root, 40)
    Insert_value(root, 60)
    Insert_value(root, 30)
    Insert_value(root, 45)
    Insert_value(root, 55)
    Insert_value(root, 70)
    print("In-Ordered traversing :")
    In_Ordered(root)
    print(" ")
    print(" ")
    print(find_value(root, 40))
    print(" ")
    print(f"Minimum value in Tree is {Min_value(root)}")
    print(" ")
    print(f"Maximum value in Tree is {Max_value(root)}")
    print(" ")
    print(f"Height of our BST is {Height(root)}")
    print(" ")
    print(" ")
    print("XXXXXXXXXXXXXXXXXXXXXXXXXXXXXX After  deleting 40 and 55 XXXXXXXXXXXXXXXXXXXXXXXXXXXXX")
    print(" ")
    Delete_Value(root, 40) # having 2 children
    Delete_Value(root, 55) # having no children
    print(" ")
    print(find_value(root, 40))
    print(" ")
    print("In-Ordered traversing of modified tree:")
    In_Ordered(root)
    print(" ")
    print(" ")
    print(f"Height of our BST is {Height(root)}")
    print(" ")
    print(" ")
    print("XXXXXXXXXXXXXXXXXXXXXXXXXXXXXX After  deleting  45 and 70 XXXXXXXXXXXXXXXXXXXXXXXXXXXXX")
    print(" ")
    Delete_Value(root, 45) # having 1 child
    Delete_Value(root, 70)
    print(" ")
    print(find_value(root, 45))
    print(" ")
    print("In-Ordered traversing of modified tree:")
    In_Ordered(root)
    print(" ")
    print(" ")
    print(f"Height of our BST is {Height(root)}")
```
