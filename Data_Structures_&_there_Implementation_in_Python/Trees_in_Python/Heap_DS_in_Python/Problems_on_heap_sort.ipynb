# Q1) Sort the given array in ascending and descending order using max and min heap ds, and find the kth smallest and largest element in it.

```python
def Min_heap(A, n, i):
    # Set smallest as root
    smallest = i
    left = 2*i  # left child would be at 2*i 
    right = 2*i + 1 # right child would be at 2*i + 1
 
    if left < n and A[left] < A[smallest]: # If left child is smaller than root
        smallest = left # Making left child as smallest element in heap
 
    if right < n and A[right] < A[smallest]: # If right child is smaller than smallest so far
        smallest = right # Making right child as smallest element in heap
 
    if smallest != i:  # If my root is not smaalest then..
        A[i],A[smallest] = A[smallest],A[i] # just swapping it with smallest element in my heap 

        # After swapping again checking if any condition is violated or not
        Min_heap(A, n, smallest) # By Recursively calling min_heap function

def Descending_HeapSort(A, n):
    for i in range(int(n / 2) - 1, -1, -1): # Building my heap from given array
        Min_heap(A, n, i)
 
    for i in range(n-1, -1, -1): # Extracting single - single element from heap

        A[0], A[i] = A[i], A[0] # Initially Moving my current root to end 
 
        Min_heap(A, i, 0) # Recursively calling min heap on my reduced heap



def Max_heap(A, n, i): 
    # Set largest as root
    largest = i 
    left = 2*i  # left child would be at 2*i  

    right = 2*i + 1  # right child would be at 2*i + 1 
   
    if (left < n and A[i] < A[left]): # if left child is greater than root

        largest = left # Make the left child as largest element in heap
  
    if (right < n and A[largest] < A[right]): # if right child is greater than root

        largest = right # Make the right child as largest element in heap
  

    if (largest != i):  # If my root is not largest then...
        A[i],A[largest] = A[largest],A[i] # just swapping it with largest element in my heap 
  
        # After swapping again checking if any condition is violated or not
        Max_heap(A, n, largest) # By Recursively calling max_heap function

def Ascending_HeapSort(A,n): 

    for i in range(n, -1, -1): # Building my heap from given array
        Max_heap(A, n, i) 
   
    for i in range(n-1, 0, -1): # Extracting single - single element from heap

        A[i], A[0] = A[0], A[i] # Moving my current root to end  

        Max_heap(A, i, 0) # Recursively calling max heap on my reduced heap

def printArray(A):
    for i in range(len(A)):
        print(A[i], end = " ")
    print()

if __name__ == "__main__":
    print(" ")
    k = int(input("Enter a number 'k': "))
    print(" ")
    a = [17, 46, 13, 21, 10, 69, 15, 4 ,11,34 ,16 ,24 ,23 ,27,25,26]
    s = str(k)
    Ascending_HeapSort(a,len(a))
    print("Ascending order of array, created by Max heap, is: ")
    print(" ")
    printArray(a)
    print(" ")
    if s[len(s)-1]=="1":
        print(f"{k}st smallest element in array is: {a[k-1]}")
    elif s[len(s)-1]=="2":
        print(f"{k}nd smallest element in array is: {a[k-1]}")
    elif s[len(s)-1]=="3":
        print(f"{k}rd smallest element in array is: {a[k-1]}")
    else:
        print(f"{k}th smallest element in array is: {a[k-1]}")
    print(" ")
    Descending_HeapSort(a,len(a))
    print("Descending order of array, created by Min heap, is: ")
    print(" ")
    printArray(a)
    print(" ")
    if s[len(s)-1]=="1":
        print(f"{k}st largest element in array is: {a[k-1]}")
    elif s[len(s)-1]=="2":
        print(f"{k}nd largest element in array is: {a[k-1]}")
    elif s[len(s)-1]=="3":
        print(f"{k}rd largest element in array is: {a[k-1]}")
    else:
        print(f"{k}th largest element in array is: {a[k-1]}")
    print(" ")
```
