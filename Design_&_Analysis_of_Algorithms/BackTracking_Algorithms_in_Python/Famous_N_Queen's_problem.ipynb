# Solving N Queen's Problem Using BackTracking

### To understand the problem follow link : 
### https://www.javatpoint.com/n-queens-problems

```python
# here i have also computed the no. of back_tracks required in each solution

class Generating_Solution:
    def Generating_matrix(self, n):
        # generating a square matrix
        arr = [['0' for i in range(n)]for j in range(n)]
        
        # creating vairable to store solution_list
        self.solution_list = []
        
        # start checking with (0,0)
        self.NQueens(arr, 0,0)

        # returning the solution list at the last
        return self.solution_list
    
    def NQueens(self,arr, row,col):
        # if row is the last row i.e.all elements are present correctly
        if row == len(arr):
            # appending that solution to sol list
            self.solution_list.append([''.join(i) for i in arr]) 
            return 
        
        # checking through all the col of given row
        for col in range(len(arr)):
            # checking if place is safe to put our Queen
            if self.Place(arr,row, col):
                # if place is safe,
                # then Placing our Queen in arr[row][col]
                arr[row][col] = '1'

                # again checking for next row and 0th col
                self.NQueens(arr, row+1,0)

                # Backtracking it to 0
                arr[row][col] = '0'

        return 
        
    def Place(self,arr, row, col):
        xrow = row
        xcol = col
        
        # checking all the row of each column
        while(row >= 0):
            if(arr[row][col] == '1'):
                return False
            row-= 1
        
        row = xrow
        col = xcol
        # Checking left upper diagonal i.e. row-- and col--
        while(row>=0 and col>= 0):
            if(arr[row][col] == '1'):
                return False
            row-= 1
            col-= 1
        
        row = xrow 
        col = xcol
        # Checking right upper diagonal i.e. row-- and col++
        while(row>=0 and col<len(arr)):
            if(arr[row][col] == '1'):
                return False
            row-= 1
            col+= 1

        # return true if current position is safe
        return True    
 
n = int(input("Enter a size of a square matrix: "))
sol = Generating_Solution()
matrix = sol.Generating_matrix(n)

if n>=4:
    print("XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX")
    for i in range(len(matrix)):
        print("Solution number:",i+1)
        for j in matrix[i]:
            print(j)
        print()
        for k in range(1):
            for l in range(k,len(matrix[i])):
                if matrix[i][k][l]=="1":
                    print("No. of BackTrack in this case is: ",l)
        print()
        print("XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX")
else:
    print(f"No solution available for matrix of size {n} x {n}")
```
